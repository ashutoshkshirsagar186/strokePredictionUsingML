```python

stroke = ['Stroke', 'No stroke']

# Manually input data for one person
user_input = {'id': 1,
              'gender': input("Enter gender (Male/Female): ").capitalize(),
              'age': float(input("Enter age: ")),
              'hypertension': int(input("Enter hypertension (0 for No, 1 for Yes): ")),
              'heart_disease': int(input("Enter heart disease (0 for No, 1 for Yes): ")),
              'ever_married': input("Enter marital status (Yes/No): ").capitalize(),
              'work_type': input("Enter work type: ").capitalize(),
              'Residence_type': input("Enter residence type (Urban/Rural): ").capitalize(),
              'avg_glucose_level': float(input("Enter average glucose level: ")),
              'bmi': float(input("Enter BMI: ")),
              'smoking_status': input("Enter smoking status: ").capitalize(),
              'stroke': None  # The actual value will be predicted
              }

# Create a DataFrame for the user input
df_user_input = pd.DataFrame([user_input])

# Handle missing values
df_user_input['bmi'].fillna(df_user_input['bmi'].mean(), inplace=True)

# Convert categorical columns to numerical using Label Encoding
le = LabelEncoder()
df_user_input['gender'] = le.fit_transform(df_user_input['gender'])
df_user_input['ever_married'] = le.fit_transform(df_user_input['ever_married'])
df_user_input['work_type'] = le.fit_transform(df_user_input['work_type'])
df_user_input['Residence_type'] = le.fit_transform(df_user_input['Residence_type'])
df_user_input['smoking_status'] = le.fit_transform(df_user_input['smoking_status'])

# Select features for the user input
X_user_input = df_user_input[['gender', 'age', 'hypertension', 'heart_disease', 'ever_married',
                               'work_type', 'Residence_type', 'avg_glucose_level', 'bmi', 'smoking_status']]

# Make predictions for the user input data using a pre-trained model (assuming 'dt' is your decision tree model)
predictions_user_input = dt.predict(X_user_input)

# Display the predicted result
print("Prediction:")
print(stroke[int(predictions_user_input[0])])
```
